# Building a Recommender using GraphLab Data Pipelines

----
In this notebook we will show a complete recommender system implemented using GraphLab Data Pipelines. This recommender example is common in many batch scenarios, where a new recommender is trained on a periodic basis, with the generated recommendations persisted to a relational database used by the web application.

The data we will use in this notebook is the same as the [Building a Recommender with Ratings Data](http://graphlab.com/learn/notebooks/recsys_explicit_rating.html) notebook, but without the exploration and prototyping parts. Instead this notebook will focus on taking the code for the pipeline and defining it using [GraphLab Data Pipelines](http://graphlab.com/products/create/docs/graphlab.deploy.html#data-pipeline).

The pipeline will contain the following tasks:

1. [Clean and transform data](#Clean-the-data)
1. [Train a Recommender model](#Train-the-model)
1. [Generate Recommendations for users](#Generate-Recommendations)
1. [Persist Recommendations to a MySQL database](#Persist-Recommendations)

Each of these tasks will be defined as a [Task](http://graphlab.com/products/create/docs/generated/graphlab.deploy.Task.html) and executed as a [Job](http://graphlab.com/products/create/docs/generated/graphlab.deploy._job.Job.html) using GraphLab Data Pipelines.

And finally, we will cover how to [Run and monitor](#Running-and-Monitoring-this-recommender) these pipelines. Remember, when using GraphLab Data Pipelines, the Tasks and Jobs created are managed objects, so they must have unique names.

__This notebook uses GraphLab Create 1.0.1__.

````python
import graphlab
````

----

## Clean the data

The first task in this pipeline will take data, clean it, and transform it into an SFrame. When a Task requires a file as input, it is passed to the framework as a parameter. In this task, the raw data is read using graphlab.SFrame.read_csv, with the file path provided as a parameter to the Task. Once the data is loaded into an SFrame, we clean it by calling dropna() on the SFrame, and then assign the cleaned SFrame as the output named 'out' for this Task.

The code that will run when the task is executed is:

````python
def clean_transform(task):
    import graphlab as gl
    sf = gl.SFrame.read_csv(task.params['csv'], delimiter='\t', column_type_hints={'rating':int})
    sf = sf.dropna()
    sf.rename({'user':'user_id', 'movie':'movie_id'})
    # to simplify this example, only keep 0.1% of the number of rows from the input data
    sf = sf.sample(0.001)
    task.outputs['out'] = sf
````


The definition for this task is as follows:

````python
clean = graphlab.deploy.Task('clean')
clean.set_params(['csv'])
clean.set_outputs(['out'])
clean.set_code(clean_transform)
clean.save()
````

----

## Train the model

Now that the data is cleaned and ready as an SFrame, we need to train a model in this recommendation system. To train the model, we need the SFrame created in the previous Task.

````python
def train_model(task):
  import graphlab as gl
  data = task.inputs['in']
  # train the model
  model = gl.recommender.create(data, user_id='user_id', item_id='movie_id', target='rating')
  task.outputs['out'] = model
````

Now to define this 'train' task in the framework. In the following Python code, the 'train' Task defines its 'in' input. The 'in' input is defined as a dependency on the 'clean' task's 'out' output. This is how to define a dependency in the Data Pipelines framework. 

It is also worth noting in the following definition the parameters and inputs are specified using a Python dictionary. Task ``set_inputs``, ``set_outputs``, and ``set_params`` methods can all take either a list of names (as strings) or a dictionary with names and values for those names. And finally, the output named 'out' is defined since it is where the trained model is assigned in the code function. And the final line saves the definition of this Task with the framework, so it can be easily loaded and logically referenced later.


````python
train = graphlab.deploy.Task('train')
train.set_inputs({'in':(clean, 'out')})
train.set_outputs(['out'])
train.set_code(train_model)
train.save()
````

----

## Generate Recommendations


With the previous task there is now a trained model that we should use for generating recommendations. With a Task now specified that trains a model, it can be improved independently from the task that generates recommendations from that model. This is a great feature of having the Tasks be separated into discrete composable parts, and supporting logical naming when refering to dependencies and tasks. Another team member can continue to iterate on the 'train' task and it can be seamlessly plugged into this pipeline when it is ready.

To generate recommendations we need the trained model to use, and the users needing recommendations. Here is the code for generating recommendations froma  trained model:

````python
def gen_recs(task):
    model = task.inputs['model']
    data = task.inputs['data']
    recs = model.recommend(data['user_id'])
    task.outputs['recs'] = recs
````

The complementary definition code for the Task:

````python
recommend = graphlab.deploy.Task('recommend')
recommend.set_inputs({'model':(train, 'out'), 'data':(clean, 'out')})
recommend.set_outputs(['recs'])
recommend.set_code(gen_recs)
recommend.save()
````

----

## Persist Recommendations


Now that recommendations have been generated, the final step in this pipeline is to save them to a relational database. The main applicaton queries this database for user recommendations as users are interacting with the application. For this task, we will use MySQL as an example, but that can easily be substituted with a different database.

The DB table needed to run this example looks like the following:
```
+----------+-------------+------+-----+---------+-------+
| Field    | Type        | Null | Key | Default | Extra |
+----------+-------------+------+-----+---------+-------+
| user_id  | varchar(50) | NO   |     | NULL    |       |
| movie_id | varchar(50) | NO   |     | NULL    |       |
| score    | float       | NO   |     | NULL    |       |
| rank     | int(8)      | NO   |     | NULL    |       |
+----------+-------------+------+-----+---------+-------+
```
To create a table in MySQL with this schema:
```
CREATE TABLE recommendations (user_id VARCHAR(50), 
                              movie_id VARCHAR(50), score FLOAT, rank INT(8));
```

````python
def persist_to_db(task):
    import mysql.connector
    from mysql.connector import errorcode

    recs = task.inputs['recs']
    dbhost = task.params['host']
    dbuser = task.params['user']
    dbpass = task.params['pass']
    dbport = task.params['port']
    dbname = task.params['name']
    dbtable = task.params['table']
    
    conn = mysql.connector.connect(host=dbhost, user=dbuser, password=dbpass, port=dbport)
    conn.database = dbname
    cursor = conn.cursor()
    # this example expects the table to be empty, minor changes here if you want to 
    # update existing users' recommendations instead.
    add_row_sql = ("INSERT INTO " + dbtable + " (user_id, movie_id, score, rank) "
                   "VALUES (%(user_id)s, %(movie_id)s, %(score)s, %(rank)s)")

    print "Begin - Writing recommendations to DB...."
    for row in recs:
        cursor.execute(add_row_sql, row)
    print "End - Writing recommendations to DB...."
    
    # commit recommendations to database
    conn.commit()
        
    # no outputs to set, since final Task in the pipeline
````

Now to define this Task:

````python
persist = graphlab.deploy.Task('persist')
persist.set_inputs({'recs':('recommend', 'recs')})
persist.set_params(['host', 'user', 'pass', 'port', 'name', 'table'])
persist.set_code(persist_to_db)

# specify any non-standard Python packages by calling set_required_packages
persist.set_required_packages(['mysql-connector-python', '--allow-external mysql-connector-python'])

persist.save()
````

**Note:** An important note about this Task is that is requires using the *mysql-connector-python* package, which is not in standard Python. Using the Data Pipelines framework, specyfing that this package is required is easily done in the Task definition. When running this task in a remote enviroment (EC2 or Hadoop) the framework will make sure this python package is installed prior to execution. See the line for ``set_required_packages``. The additional parameters are required for this package to get installed because it is not hosted by Pypi.

In order to run this pipeline locally, please install the ``mysql-connector-python`` package on your machine.

## Running and Monitoring this recommender


First, let's see what Tasks are available:

````python
graphlab.deploy.tasks
````

The output shows the tasks just defined, along with an index number we can use to load them. We will use their names when specifying the pipeline to execute. Because GraphLab Create is managing the Tasks and Jobs, each Task and Job is required to have a unique name. If you try to run this Notebook more than once it will fail because of the Tasks already existing. To remedy that, simply delete the existing task (ex. ```del graphlab.deploy.tasks['train']```) or load the existing task instead of creating a new one (ex. ```train = graphlab.deploy.tasks['train']```).

Now that the tasks are defined for this pipeline, let's compose them together to create a Job. Using the late-binding feature of the Data Pipelines framework, the parameters, inputs, and outputs that have not been specified with the Task can be specified at runtime. We will use this feature to specify the database parameters for the 'persist' task, and then raw data location for the 'clean' task.


### Save Recommendations to Disk

````python
pipeline_recs = [ 
    ('clean', {'csv':'https://s3.amazonaws.com/GraphLab-Datasets/movie_ratings/sample.small'}),
    'train',
    ('recommend', {'recs':'/tmp/movie-recs'})]

job = graphlab.deploy.job.create(pipeline_recs, name='movie-recs-file')
````

The job is started asynchronously in the background, and we can query for its status calling get_status on the Job instance returned:

````python
job.get_status()
````

To see more information about the job, print the job object:

````python
print job
````

Once the job shows status 'Completed', the recommendations SFrame is ready to load, so to verify the recommendations, let's load that SFrame in GraphLab Canvas

````python
import time
while job.get_status() == 'Running':
    time.sleep(2) # sleep for 2s while polling for job to be complete.

print job
````

````python
graphlab.canvas.set_target('ipynb') # show Canvas inline to IPython Notebook
recs = graphlab.SFrame(data='/tmp/movie-recs')
recs.show()
````


### Running in EC2 or Hadoop

Data Pipelines also supports running the same pipeline in EC2 or Hadoop YARN clusters (CDH5). In order to run this pipeline in those environments, simply add an ``environment`` parameter to ``graphlab.deploy.job.create`` API. No code needs to change, and the GraphLab Data Pipelines framework takes care of installing and configuring what is needed to run this pipeline in the specified environment.

#### To create an EC2 environment:

````python
ec2_dev = graphlab.deploy.environment.EC2('ec2_dev', 's3://my-jobs-logs/', 
                                          aws_access_key='SECRET', 
                                          aws_secret_key='SECRETAGAIN')
ec2_dev.save()
````

#### To create a Hadoop environment:

````python
hd = graphlab.deploy.environment.Hadoop('hd', config_dir='~/yarn', memory_mb=16384)
hd.save()
````

The ```config_dir``` parameter should point to the directory that contains yarn-site.xml. The ```memory_mb``` parameter indicates how much memory to request from YARN when scheduling the job.
